```python
import random

session=['04.13(월) 슬롯1', '04.16(목) 슬롯2', '04.16(목) 슬롯3','04.20(월) 슬롯4']
per_session=[4,4,5,5]

teams=[]
missing=[4]
for i in range(1,20):
  if i not in missing: teams.append(i)
print(teams)
random.shuffle(teams)
print(teams)
print(f'total {len(teams)}팀')

# 한슬롯당 5팀씩 진행하는 것으로 가정

count=0
for i in range(len(session)):
  print(session[i], end='')
  for j in range(per_session[i]):
    if count == len(teams): break
    print(f' {teams[count]}', end='')
    count+=1
  print()
```
